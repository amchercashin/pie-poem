# Ежи Лец - цитаты, сборка модели

```python
def read_quotes(file_name: str) -> list:
    quotes = []
    with open(file_name, encoding='utf-8') as file:
        lines = file.readlines()
        quote = ""
        for line in lines:
            quote = line.strip() 
            if quote:
                quotes.append(quote)
    return quotes
```

```python
read_quotes('./data/Lecs.txt')
```

```python
from poems_model import PoemsModel

pm = PoemsModel()

poems_file = './data/Lecs.txt'
w2v_file = 'c:/data/ruscorpora_1_300_10.bin.gz'

pm.compile(poems_file, w2v_file, poems_reader=read_quotes)
```
